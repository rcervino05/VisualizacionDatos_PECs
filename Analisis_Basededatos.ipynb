## Estudio de los datos

A continuación, se procederá a analizar el conjunto de datos obtenidos de la web anteriormente mencionada. En este apartado se estudiarán las variables de la muestra, su distibución o valores nulos, entre otros.


```python
import pandas as pd
import numpy as np

#Se carga el fichero con los datos
data = pd.read_csv("pax_all_agreements_data.csv", sep = ",")

#Se inspeccionan las 5 primeras líneas
data.head(5)
```

Estudiando la salida anterior, se aprecia que existen más de 260 atributos, entre los que se encuentran variables numéricas y categóricas. Debido al gran volumen de variables, no se pueden explicar todas, con lo que se invita a investigar acerca de ellas a partir del documento que se puede encontrar en el siguiente fichero  https://www.peaceagreements.org/files/PA_X_codebook_Version2_June_2019.pdf.

Entre las variables principales se pueden ver Con, lugar donde se originan los conflictos, o Contp, el tipo de conflicto.

```python
#Se obtiene una descripción de las variables numéricas
data.describe(include=[np.number])
```

En esta ejecución se pueden apreciar los estadísticos principales de los atributos, como el número de observaciones del dataset, 1782, la media, desviación típica, mínimo, máximo y percentiles.

```python
#Se estudian los valores nulos
data.isnull().sum()
```
